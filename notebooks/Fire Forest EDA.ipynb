## EDA

```python
# Changing Numeric 
for col in df.columns[:-1]:
    df[col] = pd.to_numeric(df[col] , errors='coerce')
```

```python
# Checking NUll
df.isnull().sum()
```

```python
# dropping missing values as it is not greater in number 
df = df.dropna()
```

```python
df['Classes  '].unique()
```

```python
# Mapping classes columns 
df['Classes  '] = df['Classes  '].astype(str).str.replace(r'\s+' , '',regex = True).str.strip().map({'notfire' : 0 , 'fire':1})
```

```python
## dropping unnecessary columns 
df = df.drop(['day','month','year'] , axis= 1)
```
